# Visualization 1:
# Distribution of titles by type

```python
# Set style for the plot
sns.set_theme(style="whitegrid")  # Use the whitegrid style for the plots for better readability

# Plot the count of each type
plt.figure(figsize=(8, 5))  # Set the figure size for the plot
sns.countplot(data=data, x="type", palette="viridis")  # Create a count plot for the 'type' column using the viridis color palette
plt.title("Distribution of Titles by Type", fontsize=14)  # Set the title of the plot with font size 14
plt.xlabel("Type", fontsize=12)  # Set the x-axis label with font size 12
plt.ylabel("Count", fontsize=12)  # Set the y-axis label with font size 12
plt.xticks(fontsize=10)  # Set the font size of x-axis tick labels to 10
plt.yticks(fontsize=10)  # Set the font size of y-axis tick labels to 10
plt.show()  # Display the plot

# Insight: Count the number of titles by type
type_distribution = data['type'].value_counts()  # Count the occurrences of each type (Movie or TV Show)
type_distribution  # Display the distribution of types
```

# Insights:-
* Plot Customization: The code sets the plot style to whitegrid for better readability, which enhances the visual appeal of the plots.

* Figure Size: It specifies the figure size, ensuring that the plot is appropriately sized for better viewing.

* Count Plot: The countplot function is used to visualize the distribution of titles by their type (Movie or TV Show). The palette="viridis" applies a color palette to the plot.

* Plot Titles and Labels: The plot is given a title, and the x and y axes are labeled with specific font sizes to make the plot more informative and readable.

* Tick Labels: The font sizes of the tick labels on both axes are set for improved readability.

* Distribution Count: The code calculates the distribution of titles by type using value_counts(), which provides a count of each type (Movies and TV Shows).

# Visualization 2:
# Number of titles added per year

```python
# Convert the 'date_added' column to datetime format
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extract the year from the 'date_added' column
df['year_added'] = df['date_added'].dt.year

# Plot the number of titles added per year
plt.figure(figsize=(12, 6))  # Set the figure size
sns.countplot(data=df, x='year_added', palette='viridis')  # Create a count plot
plt.title('Number of Titles Added Per Year')  # Set the title of the plot
plt.xlabel('Year Added')  # Set the x-axis label
plt.ylabel('Count')  # Set the y-axis label
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()  # Display the plot
```

# Insight:-
* This visualization shows the number of titles added to the database each year.
* It helps to understand trends over time, such as increases or decreases in the number of titles added annually.

---

# Visualization 3:
# Distribution of movies by duration

```python
# Filter the DataFrame to include only movies
movies_df = df[df['type'] == 'Movie']

# Convert 'duration' to numeric, removing 'min' from the string
movies_df['duration'] = movies_df['duration'].str.replace(' min', '').astype(float)

# Plot the distribution of movie durations
plt.figure(figsize=(12, 6))  # Set the figure size
sns.histplot(data=movies_df, x='duration', bins=20, kde=True, palette='viridis')  # Create a histogram plot
plt.title('Distribution of Movie Durations')  # Set the title of the plot
plt.xlabel('Duration (minutes)')  # Set the x-axis label
plt.ylabel('Count')  # Set the y-axis label
plt.show()  # Display the plot
```

# Insight:-
* This visualization shows the distribution of movie durations.

* It provides insights into the typical length of movies and identifies any outliers in terms of duration.

---

# Visualization 4:
# Top countries by the number of titles released

```python
# Count the number of titles by country
country_counts = df['country'].value_counts().head(10)  # Get top 10 countries

# Plot the top countries by number of titles
plt.figure(figsize=(12, 6))  # Set the figure size

# Create a bar plot
country_counts.plot(kind='bar', colormap='viridis')  # Use the colormap parameter

plt.title('Top Countries by Number of Titles')  # Set the title of the plot
plt.xlabel('Country')  # Set the x-axis label
plt.ylabel('Count')  # Set the y-axis label
plt.show()  # Display the plot
```

# Insight:-
* This visualization highlights the top 10 countries by the number of titles they have produced.

* It allows us to see which countries are major contributors to the dataset.

---

# Visualization 5:
# Number of titles released per genre

```python
# Split the 'listed_in' column to separate genres and stack them
genres = df['listed_in'].str.split(', ', expand=True).stack().value_counts()

# Plot the number of titles per genre
plt.figure(figsize=(14, 8))  # Set the figure size
genres.plot(kind='bar')  # Use default colors
plt.title('Number of Titles per Genre')  # Set the title of the plot
plt.xlabel('Genre')  # Set the x-axis label
plt.ylabel('Count')  # Set the y-axis label
plt.show()  # Display the plot
```

# Insight:-
* This visualization shows the number of titles per genre.

* It provides insights into the most popular genres in the dataset.

---

# Visualization 6:
# Heatmap of Ratings by type

```python
# Create a pivot table for the heatmap
heatmap_data = df.pivot_table(index='rating', columns='type', aggfunc='size', fill_value=0)

# Plot the heatmap of ratings by type
plt.figure(figsize=(12, 8))  # Set the figure size
sns.heatmap(heatmap_data, annot=True, fmt='d')  # Create a heatmap
plt.title('Heatmap of Ratings by Type')  # Set the title of the plot
plt.xlabel('Type')  # Set the x-axis label
plt.ylabel('Rating')  # Set the y-axis label
plt.show()  # Display the plot
```

# Insight:-
* This visualization shows a heatmap of the number of titles by rating and type (Movie or TV Show).

* It helps to understand the distribution of ratings across different types of titles and identify any patterns.
